```python
import pandas as pd
import os
import sqlite3
import numpy as np 
from lets_plot import*
LetsPlot.setup_html()
import matplotlib.pyplot as plt
```

```python
filename='D:/me2024/Meesho Analysis/Data/Raw/meesho_df.csv'
df=pd.read_csv(filename)
```

```python
labels = ['Cotton', 'Lycra', 'crepe', 'Georgette','Rayon','Polyester']
sizes = [50, 30, 20, 10,30,20]  # Corresponding sizes for each category
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue','blue','gray']  # Custom colors
explode = (0.1, 0, 0, 0,0,0)  # "Explode" the first slice (optional)

# Create a pie chart
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)

# Set the aspect ratio of the chart to be equal, so the pie chart is drawn as a circle.
plt.axis('equal')

# Add a title
plt.title('Fabrics')

# Display the pie chart
plt.show()
```

```python
ggplot (df, aes(x='fabrics',y='ratings')) + \
    geom_bar(star='identify',fill='blue') + \
    ggtitle('Number of fabrics by Ratings') + \
    xlab('Fabrics') + \
    ylab('Ratings') + \
    theme (axis_text_x=element_text(angle=45, hjust=1)
    )
```

```python
ggplot(df,aes(x='sizes', y='prices')) + \
    geom_line(color='red', size=2) + \
    ggtitle('Sizes wise Prices') + \
    xlab('Size') + \
    ylab('Price')
```

```python
import pandas as pd
import plotly.express as px

# Assuming columns: 'Category', 'Subcategory', 'Sales'
fig = px.sunburst(df, path=['patterns', 'fabrics'], values='prices', title="Prices wise Patterns and Fabrics")
fig.show()

```

```python
import plotly.express as px

# Assuming columns: 'Category', 'Product', 'Revenue'
fig = px.treemap(df, path=['sizes', 'patterns'], values='ratings', title="Ratings Wise Sizes and Patterns")
fig.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Plot hexbin
plt.figure(figsize=(8, 6))
plt.hexbin(df['ratings'], df['prices'], gridsize=30, cmap='coolwarm', mincnt=1)
plt.colorbar(label="Count of Products")
plt.xlabel("")
plt.ylabel("Sales")
plt.title("Prices VS Ratings")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Group by category
category_sales = df.groupby('names')['ratings'].sum()

# Plot
plt.figure(figsize=(7, 7))
plt.pie(category_sales, labels=category_sales.index, autopct='%1.1f%%', startangle=140, wedgeprops=dict(width=0.3))
plt.title("Names wise Ratings")
plt.show()
```
