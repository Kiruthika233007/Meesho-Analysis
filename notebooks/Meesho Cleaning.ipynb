```python
import pandas as pd
import os
import sqlite3
import numpy as np 
```

```python
filename='D:/me2024/Meesho Analysis/Data/Raw/meesho_1.csv'
df=pd.read_csv(filename)
```

```python
df
```

```python
columns_int = ['prices', 'net_quantities', 'ratings']
df['prices'] = df['prices'].replace('₹', '', regex=True)  # Remove rupee symbol
df[columns_int] = df[columns_int].apply(pd.to_numeric, errors='coerce')  # Convert columns to numeric, invalid parsing will be set to NaN

# Now rename and convert columns to Int64
df.rename(columns={'Unnamed: 0': 'product_id',
                   'delivery_':'delivary'}, inplace=True)

df[columns_int] = df[columns_int].astype('Int64', errors='ignore')
```

```python
df
```

```python
conn = sqlite3.connect(os.path.join('../data', 'meesho.db'))
```

```python
df.to_sql('Meesho_products' ,conn, if_exists='replace', index=False)
```

```python

meesho_product=conn.execute('''
    select * from Meesho_products
''').fetchall()
```

```python
meesho_df=df = pd.DataFrame(meesho_product, columns=['product_id','names','sizes','prices', 'fabrics', 'sleeves','net_quantities','patterns', 'ratings','delivery'])
```
